HIPÓTESIS BÁSICAS:

Todos estmos rodeados de tópicos, pero en cuanto al planeta hay unos cuantos que se requiere analzar.

 - La población rural desminuye año tras año.

 - Los países son todos iguales, porque son concentración de poder sobre un territorio.

 - Las tierra agricolas aumentan año tras año.

 - Los bosque disminuyen año tras año.

 - La producción de limentos no es suficiente para alimentarnos a todos.

PREGUNTAS y RESPUESTAS:

A) EN CUANTO A LA POBLACIÓN:

  0 - ¿Cuál es la TENDENCIA general de la población mundial?

     - La población mundial se ha incrementado década tras década desde hace, por lo menos, 50 años.

     - Hay una tendencia al incremento poblacional en la mayoría de los países, salvo en unas pocas decenas de países que coinciden con los tradicionalmente más desarrollados, o con aquellos donde ha habido guerras u otras crisis.

  1 - ¿Hay países cuya población decrece?, ¿Cómo analizamos las TASAS DE CRECIMIENTO de la población?

     - Hay 46 PAÍSES cuya población decrece año tras año, se concentran en Europa, y curiosamente los países con mayor y menor tasa de crecimiento poblacional son también europeos.

     - Analizando las tasas de crecimiento por CONTINENTES vemos que todos los continente tienen tasas positivas. Además hay una mayoría de países que crecen por encima de la tasa de crecimiento media mundial.

     - Tener en cuenta que el ritmo al que todos los continentes crecen está muy cerca del ritmo de crecimiento nulo, aunque unos crezcan más que otros. 

  2 - ¿Está la población mundial homogeneamente distribuída en el mundo?

     - Más del 45 % de la población mundial se concentra en 5 PAÍSES, 4 de ellos asíaticos, siendo China e India los más populosos con diferencia superando ambos los 1.300 millones de habitantes.

     - En Asia viven casi 2/3 partes de la poblacion mundial, mientras que en Oceanía el 0.5 %. 

     - El continente más densamente poblado es Asia, siendo el 2º más extenso y populoso, mientras que Europa esta densamente poblado
    
     - ÁFRICA es el continente que más rápido crece y su población supone casi 1/5 de la población mundial.
   
     - En cuanto a DENSIDAD DE POBLACIÓN Asia es el que más concentración de personas tiene por km2, junto con Europa, pero es relevante la poca densidad poblacional de Sudamérica, siendo éste de una extensión considerable. Sin embargo, solo Asia y África están por encima de la media.

    - El RATIO DE POBLACIÓN demuestra la hipótesis inicial, mostrando que en los últimos 50 años la población total se ha triplicado. 
 

  3 - ¿Qué diferencia de SUPERFICIE hay entre los países?

     - Los 5 PAÍSES más extensos del mundo ocupan el 40% de la superficie mundial, siendo RUSIA el de mayor superficie mundial.

     - Hay países tan pequeños como una pequeña ciudad de cualquier otro país.

     - Los 2 CONTINENTES más extensos son África y Asia, seguido de cerca por Sudamérica, siendo Europa el más pequeño en superficie por debajo de Oceanía. 

  4 - ¿Qué % de la población mundial representa la POBLACIÓN RURAL y qué cambios se han producido en los últimos 50 años?

    - En 2016 representaba el 40 % de la población total.

    - Se ha visto un total descenso de la población rural en casi todos los países del mundo, alrededor del 25 %. Los más afectados han sido los países africanos. Entre los que han incrementado ese % son mayoría americanos, y entre los que se han mantenido estables son centroasíaticos y europeos.

  5 - ¿Qué representan los BOSQUES en el mundo?, ¿Qué cambios han sufrido?, ¿Cuánta superficie representa lo CULTIVADO?

    - A día de hoy los bosques representan el 34% de la supreficie terrestre, 2/3 son bosque.

    - La tierra cultivada representa alrededor del 4.5% de la superficie terrestre, y la cultivable un 15%.

    - Hasta el año 2000 se produjo un aumento paulatino de las tierra de cultivo, pero después de esoe año han ido disminuyendo progresivamente.

B) EN CUANTO A LA PRODUCCIÓN DE ALIMENTOS:

  1 - ¿Cuál es la TENDENCIA MUNDIAL en alimentación?

    - Hay una clara tendencia al aumento de producción de alimentos año tras año, sobre todo en relación a la destinada al consumo humano.

  2 - ¿Qué tipo de PRODUCTOS ALIMENTARIOS se producen?

    - Hay determinados produtos, unos 10, que superan con creces al resto en miles de toneladas de alimento producidas.

    - Los productos de mayor producción, cereales, fruta, maíz, leche y mantequilla, patatas, arroz, tubérculos, vegetales, trigo.

  3 - ¿Cuál es la situación y evolución en cuanto a la ALIMENTACIÓN HUMANA Y ANIMAL por CONTINENTES?

    - Asia es el mayor productor de alimentación humana, y Europa de alimentación animal. Sin embargo, hace 50 lo fue Europa en ambos casos.

    - África ha igualado a Europa en la actualidad en alimentación humana.

    - En cuanto a la relación entre población y producción de alimentos mundiales por continente Asía y Oceanía están muy lejos de la media, mientras África está justo en la media de ambas variables. 

  4 - ¿Cuál es la situación y evolución en cuanto a la ALIMENTACIÓN HUMANA Y ANIMAL por PAÍSES?

    - El mayor productor de alimentos mundial es India, sin embargo, no es uno de los mayores productores de comida para consumo animal.

    - El máximo productor de alimentos para humanos es India, y el de alimentos para animales Brazil.

    - Hay 2 países europeos, Ucrania y Alemania, que son unos de los mayores productores de comida para consumo animal, pero no para consumo humano.

    - Los 5 máximos produtores de alimentos producen 1/3 de la producción mundial.

  5 - Conlusión general.

  - Hay una tendencia mundial al aumento año tras año, tanto de producción de alimentos humana, como animal.

  - Hay una clara correlación entre aumento de la población y aumento de la producción de alimentos global. El ratio de Población y el de alimentos son prácticamente iguales.

   - En 1990 y 2008 hubo un decrecimiento en la producción de alimentos mundiales, que frenó de forma puntual la tendencia a incrementarse año tras año, probablemente por sendas crisis que se produjeron.
